#### Project Report

The act report displays the insights and visualiztion produced from the wrangled data and which was stored into a single dataframe and saved as a csv file. 

The focus of the project is to gather data from different sources and generate meaninful insights using analysis and visualization.

#### Project Goal

This project aims to teach the user how to carry out analysis and visualise data from the wrangled data created in the first part of the project. Three (3) insights are to be produced for the analysis and visualisations are to be created to support the insights. 

#### Insights

The combination of the three datasets stored into a csv file as **'twitter_archive_master.csv** led to the generation of  three (3) insights listed below 

1. The most popular dog stage is 'pupper' with the least being 'floofer'.

2. There is a direct correlation between retweet count and favorite count. This means the higher the retweets, the higher the favourites count.

3. The highest occuring image number according to the prediction. 

#### Visualizations

**1. The most popular dog stage is 'pupper' with the least being 'floofer'**

This was created using the **countplot()** function in creating a bar chart that shows the count of each dog stage. The dog stages that had two stages were ignored (i.e doggo,pupper). The visualizations aims to show the popularity on one single stage only.



**2. There is a direct correlation between retweet count and favorite count. This means the higher the retweets, the higher the favourites count.**

The 2nd visualisation aims to show the relationship between the two variables, retweet_count and favourite_count. Using the jointplot()function, a scatterplot was created between the 2 variables and contributed to the insight provided that "There is a direct correlation between retweet count and favorite count. The higher the retweets, the higher the favourites count."



**3. The highest occuring image number according to the prediction**

The 3rd visualisation was created to show the highest occuring image number according to each tweet's most confident prediction. This showed that the highest occuring was 1 with a percentage distribution of about 86%. This visualization was created using the seaborn's countplot() function.
